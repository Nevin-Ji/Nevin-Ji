```python
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
```

```python
gzPath = 'E:\PolyU\gis\Dissertation\Data\广州市'
demandPath = 'E:\PolyU\gis\Dissertation\Data\GZ_demand'
```

```python
#列出所有demand dataset
def ListAllSta(cityPath):
    dayPath = os.listdir(cityPath)
    allDayPath = []
    for day in dayPath:
        if 'zip' not in day and 'csv' not in day:
            allDayPath.append(os.path.join(cityPath, day))
    allPath = []
    for curPath in allDayPath:
        allTime = os.listdir(curPath)
        for curTime in allTime:
            if 'csv' in curTime :
                allPath.append(os.path.join(curPath,curTime))
    demandValueList = [s for s in allPath if 'totalStaCharging' in s]
    return demandValueList
```

```python
def demandCal(stademand):
    df1 = pd.read_csv(stademand)
    a = df1['id'].unique()
    df2 = pd.DataFrame(a)
    df2.columns = ['id']
    demandList = ['demand_value']
    df2 = pd.concat([df2, pd.DataFrame(columns = demandList)])
#     print(df3)
# 计算每一个充电站的每日需求量
    cnt = 0
    for i in a:
        #     print(i)
        df3 = df1[df1['id'] == i]
        b = df3['usedStaElec'].sum()
        df2.iloc[cnt,1] = b
#         print(df2)
        cnt = cnt+1
    print(df2)
    df2.to_csv('E:\\PolyU\\gis\\Dissertation\\Data\\GZ_demand\\' + stademand.split('\\')[-2] + '.csv',encoding = 'gbk')
```

```python
for i in ListAllSta(gzPath):
    print(i)
    demandCal(i)
```
