```python
def TotalChargingCSVFile(cityPath):
    keywords = '_total' and 'Charging'
    allPath = []
    paths = os.walk(cityPath)
    for root,dirs,files in paths:
        for file in files:
            if keywords in file:
                if file[:1] is '2'and 'Sta' not in file:
#                     print(file)
                      allPath.append(os.path.join(root,file)) #带路径的文件名
    
    return allPath
```

```python
def TotalStaChargingCSVFile(cityPath):
    keywords = '_totalStaCharging'
    allPath = []
    paths = os.walk(cityPath)
    for root,dirs,files in paths:
        for file in files:
            if keywords in file:
                if file[:1] is '2':
#                     print(file)
                    allPath.append(os.path.join(root,file)) #带路径的文件名
    
    return allPath
```
