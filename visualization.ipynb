```python
import pandas as pd
import folium
import seaborn as sns
import matplotlib.colors as colors
from folium.plugins import MarkerCluster

# 读取表格数据
df = pd.read_csv('E:\\PolyU\\gis\\Dissertation\\Data\\dispatching_results050_new1_3.csv')

# 将经纬度数据转换成float类型
df[['blat', 'blon']] = df[[ 'latitude', 'longitude']].astype(float)
df[['flat', 'flon']] = df[['tgt_latitude', 'tgt_longitude']].astype(float)

# # 选择一个颜色映射方案
# cmap = sns.color_palette("Paired", n_colors=len(df['tgt_id'].unique()))

# 创建地图
m = folium.Map(location=[df['blat'].mean(), df['blon'].mean()], zoom_start=10,control_scale=True)

# 添加英文版OpenStreetMap地图图块
folium.TileLayer(tiles='https://server.arcgisonline.com/ArcGIS/rest/services/World_Street_Map/MapServer/tile/{z}/{y}/{x}', attr='OpenStreetMap', name='OpenStreetMap').add_to(m)

# # 定义弹出框内容函数
# def create_popup_content(row):
#     content = f"""
#         <h4>电量接收者 {row['id']}</h4>
#         <ul>
#             <li>位置: 纬度 {row['blat']:.5f}, 经度 {row['blon']:.5f}</li>
#         </ul>
#     """
#     return content

# 添加电量接收者Marker
for index, row in df.iterrows():
    folium.CircleMarker(
        location=[row['blat'], row['blon']],
        radius=3,
        fill=True,
        fill_opacity=0.5,
        color_opacity=0.5,                     #!
        color='red',
        fill_color='red',
    ).add_to(m)

# # 生成不同颜色的字典
# colors_dict = {}
# for i, tgt_id in enumerate(df['tgt_id'].unique()):
#     colors_dict[tgt_id] = colors.rgb2hex(tuple(cmap[i][:3]))


# 添加电量提供者Marker和Arrow marker
marker_cluster2 = MarkerCluster().add_to(m)
for index, row in df.iterrows():
    if not pd.isna(row['tgt_id']):
        # 只有电量提供者非空才绘制箭头        
#         color = colors_dict[row['tgt_id']]
        folium.CircleMarker(
        location=[row['flat'], row['flon']],
        radius=3,
        fill=True,
        fill_opacity=1.0,
        color='green',
        fill_color='green',
        ).add_to(m)
        folium.PolyLine(
            locations=[[row['flat'], row['flon']], [row['blat'], row['blon']]],
            weight=2,
            opacity=0.5,
            color='blue',
            arrowhead=True,
            tooltip=str(row['tgt_id']) + ' to ' + str(row['id'])
        ).add_to(m)

# 保存并可视化地图
m.save('C:/Users/Nevin Ji/Desktop/map050_v3.html')
m
```
