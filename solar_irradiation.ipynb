```python
# for循环处理计数
countList = []
cnt = 0
for i in range(len(df2['total_solar_radiation'])):
    s = eval(pd.to_numeric(df1.iloc[cnt,1], errors='ignore'))
    t =pd.to_numeric(s, errors='ignore')
    x = np.count_nonzero(t, axis=0)
    countList.append(x)
    cnt =cnt+1
print(countList)
```

```python
#计算分配后的sr值 
df2['count'] = countList
df2['sr_mean_value'] = df2.loc[:,'total_solar_radiation']/df2.loc[:,'count']
df2
```
